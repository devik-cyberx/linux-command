1. what is linux?

2. What is the difference between Hard Link & Soft Link?

Ans- Hard Link = Duplicate name for same file content.

   - Soft Link = Shortcut (just points to the original file).

3. What is a Kernel in Linux?

Ans- The Kernel is the core part of the Linux operating system.
It works like a bridge between your hardware (CPU, memory, etc.) and software (apps, programs).

  💡 Simple Example:

     When you play a video or save a file,
     the kernel tells your hardware what to do.
     It manages memory, processes, files, and devices.

5. I What is the ‘grep’ command used for in Linux?

Ans- grep is a command used to search for text in files. It finds lines that match a word or pattern.
  
     Example:

          grep "hello" file.txt
🔹 This will search for the word "hello" in file.txt.

7.  Step1: Create directory /tmp/bg as root user and create files inside it.

 Step2: “abhi” should be the owner of the directory. He should be able to create files and delete files inside

 the directory and also he should be able to add content to all files inside the directory

 Ans-   
      step1. Create directory /tmp/bg and files inside

          sudo mkdir /tmp/bg
          cd /tmp/bg
          sudo touch file{1..3}.txt

 step2, Change owner of directory to user abhi

      sudo adduser abhi

      sudo chown -R abhi tmp/bg

Step 3: Set permissions so that abhi can

      sudo chmod -R u+rwX /tmp/bg
    

switch to abhi

    sudo su - abhi
